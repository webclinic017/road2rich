```python
def nor(y):
    return (y-np.min(y))/(np.max(y)-np.min(y))

df1 = df1[df1['date']>='2015-05-17']
df2 = df2[df2['date']>='2015-05-17']


x1,y1 = df1['date'],df1['close'].to_numpy()
y1 = nor(y1)
x2,y2 = df2['date'],df2['close'].to_numpy()
y2 = nor(y2)
x3,y3 = df3['日期'],df3['收盘'].to_numpy()
y3 = nor(y3)
plt.plot(x1,y1,'r',label='all')
plt.plot(x2,y2,'b',label='bank')
plt.plot(x3,y3,'g',label='601166')
plt.legend() # 显示图例
```
