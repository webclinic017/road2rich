```python
import akshare as ak
import numpy as np
```

```python
# 板块限制
stock_area = {
    "600":"上交所主板",
    "601":"上交所主板",
    "603":"上交所中小板",
    "605":"上交所主板",
    "000":"深交所主板",
    "001":"深交所主板",
    "002":"深交所中小板",
    "003":"深交所主板",
    "300":"创业板",
    "301":"创业板",
    "688":"科创板",
    "689":"科创板",
    "430":"新三板",
    "8":"北交所"
}
```

```python
stock_zh_a_spot_em_df = ak.stock_zh_a_spot_em()
```

```python
def bms(x):
    if x>1e8:
        return "大"
    elif x<5e7:
        return "小"
    else:
        return "中"

def get_bankuai(x):
    start_area = str(x)[:3]

    area = stock_area.get(start_area,'NOTSURE')
    if area =='NOTSURE':
        one_digital = str(x)[:1]
        area = stock_area.get(one_digital,'NOTSURE')
    return area


stock_zh_a_spot_em_df['流通股本'] = stock_zh_a_spot_em_df.apply(lambda row:row['流通市值']/row['最新价'],axis=1)
stock_zh_a_spot_em_df['大中小盘'] = stock_zh_a_spot_em_df['流通股本'].apply(lambda x:bms(x))
stock_zh_a_spot_em_df['交易板块'] = stock_zh_a_spot_em_df['代码'].apply(lambda x:get_bankuai(x))
stock = stock_zh_a_spot_em_df
stock_zh_a_spot_em_df.head()
```
